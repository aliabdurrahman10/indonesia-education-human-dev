## Definisi
Indeks Pembangunan Manusia (IPM) adalah indikator komposit yang digunakan untuk mengukur capaian pembangunan manusia di suatu wilayah.  
IPM dihitung berdasarkan tiga dimensi dasar pembangunan manusia:
1. **Kesehatan**: umur panjang dan hidup sehat (diukur dengan harapan hidup saat lahir).  
2. **Pendidikan**: pengetahuan (diukur dengan rata-rata lama sekolah dan harapan lama sekolah).  
3. **Standar hidup layak**: diukur dengan pengeluaran per kapita yang disesuaikan.  

Semakin tinggi nilai IPM, semakin baik kualitas pembangunan manusia di wilayah tersebut.

### **1. σ-Convergence**

Mengukur apakah standar deviasi antar provinsi menurun dari waktu ke waktu.

```python
# Hitung standar deviasi tiap tahun
sigma = df.groupby("tahun")["rata_rata_lama_sekolah"].std().reset_index(name="std_dev")

# Plot tren sigma convergence
plt.plot(sigma["tahun"], sigma["std_dev"], marker="o")
plt.xlabel("Tahun")
plt.ylabel("Standar Deviasi RLS antar Provinsi")
plt.title("σ-convergence Rata-Rata Lama Sekolah (2010–2024)")
plt.show()
```

#### Interpretasi: kalau grafik standar deviasi cenderung menurun, artinya terjadi σ-convergence (gap antar provinsi makin kecil).

### N2. **β-Convergence**

Regresi pertumbuhan terhadap level awal.
Misalnya: hitung pertumbuhan 2010–2024 → regresikan terhadap level awal 2010.

```python
# Ambil data awal (2010) dan akhir (2024)
df_2010 = df[df["tahun"] == 2010][["nama_provinsi", "rata_rata_lama_sekolah"]]
df_2024 = df[df["tahun"] == 2024][["nama_provinsi", "rata_rata_lama_sekolah"]]

# Merge
merged = pd.merge(df_2010, df_2024, on="nama_provinsi", suffixes=("_2010", "_2024"))

# Hitung pertumbuhan tahunan rata-rata (CAGR)
merged["growth"] = (np.log(merged["rata_rata_lama_sekolah_2024"]) - 
                    np.log(merged["rata_rata_lama_sekolah_2010"])) / (2024 - 2010)

# Regressi: growth ~ level awal
X = sm.add_constant(merged["rata_rata_lama_sekolah_2010"])
y = merged["growth"]
model = sm.OLS(y, X).fit()

print(model.summary())
```

### Interpretasi:

Jika koefisien pada rata_rata_lama_sekolah_2010 negatif signifikan → ada β-convergence.

Artinya provinsi dengan rata-rata lama sekolah rendah di awal tumbuh lebih cepat dibanding yang tinggi.
