# Analisis Konvergensi Pendidikan Antar Provinsi di Indonesia

Tujuan dari analisis konvergensi ini adalah untuk mengetahui apakah pembangunan pendidikan di Indonesia berlangsung secara merata antar provinsi. Dengan melihat tren ketimpangan dan kecepatan pertumbuhan rata-rata lama sekolah, kita bisa menilai apakah provinsi-provinsi yang tertinggal berhasil mengejar provinsi yang lebih maju. Hasil ini penting untuk mengevaluasi efektivitas kebijakan pendidikan nasional, mengidentifikasi wilayah yang masih tertinggal, dan memberikan bukti empiris bagi pengambilan keputusan agar pemerataan pendidikan benar-benar tercapai.

### **1. σ-Convergence**

Mengukur apakah standar deviasi antar provinsi menurun dari waktu ke waktu.

```python
# === σ-Convergence: Standar deviasi antar provinsi dari waktu ke waktu ===
import pandas as pd
import matplotlib.pyplot as plt

# 1) Hitung standar deviasi tiap tahun
sigma = (
    df.groupby("tahun")["rata_rata_lama_sekolah"]
      .std()
      .reset_index(name="std_dev")
)

# 2) Plot tren sigma convergence
plt.figure(figsize=(8,5))
plt.plot(sigma["tahun"], sigma["std_dev"], marker="o", linestyle="-")
plt.xlabel("Tahun")
plt.ylabel("Standar Deviasi RLS antar Provinsi")
plt.title("σ-Convergence Rata-Rata Lama Sekolah (2010–2024)")
plt.grid(True)
plt.show()

# 3) Interpretasi cepat
if sigma["std_dev"].iloc[-1] < sigma["std_dev"].iloc[0]:
    print("HASIL: Gap antar provinsi makin kecil (indikasi σ-convergence).")
else:
    print("❕ Gap antar provinsi tidak mengecil (tidak ada σ-convergence).")
```

**Apa yang Ditunjukkan Grafik?**

- Grafik menggambarkan **standar deviasi rata-rata lama sekolah antar provinsi** setiap tahun.  
- Di awal (2010–2012), standar deviasi cukup tinggi (sekitar 1.6).  
  → Artinya, ada perbedaan cukup lebar antar provinsi dalam hal lama sekolah.  
- Setelah 2013, terlihat penurunan tajam → standar deviasi stabil di kisaran 0.9–1.0.  
- Dari 2014 sampai 2023, gap antar provinsi relatif kecil dan stabil.  
- Namun, tahun 2024 ada sedikit kenaikan kembali (sekitar 1.2).  
  → Ini bisa jadi karena data terbaru menunjukkan adanya provinsi baru atau dinamika khusus di wilayah tertentu.


**Beberapa Insight yang Bisa Diambil**

1. **Konvergensi Pendidikan Terjadi**  
   Penurunan standar deviasi menunjukkan bahwa ketimpangan antar provinsi dalam rata-rata lama sekolah makin mengecil.  
   Dengan kata lain, provinsi-provinsi yang dulu tertinggal mulai mengejar ketertinggalannya.

2. **Stabilitas Relatif setelah 2013**  
   Setelah 2013, variasi antar provinsi jauh lebih sempit.  
   Hal ini bisa mencerminkan kebijakan nasional (misalnya *Wajib Belajar 12 Tahun*, BOS, pemerataan akses sekolah) yang mulai berdampak.

3. **Kenaikan Kecil di 2024**  
   Kenaikan di ujung bisa mengindikasikan adanya provinsi yang datanya “ekstrim” dibanding yang lain.  
   Bisa juga karena adanya pemekaran wilayah (misalnya Papua) sehingga perbedaan statistik antar provinsi jadi lebih jelas.


**Kenapa pakai metode σ-Convergence?**

σ-Convergence adalah metode yang melihat ketimpangan dengan standar deviasi atau varians.

Logikanya sederhana:

- Kalau standar deviasi antar wilayah menurun, artinya perbedaan makin kecil → ada konvergensi.  
- Kalau standar deviasi tetap atau naik, artinya gap antar wilayah tidak mengecil.  

Cocok dipakai untuk data panel antar wilayah dan tahun, karena fokusnya memang ke “apakah ketimpangan antar wilayah makin sempit”.


**Apa relevansi pemilihan metode dengan tujuan penelitian ini?**

Pemerintah sering bicara soal pemerataan pembangunan.  

Di bidang pendidikan, bukan cuma **rata-rata nasional** yang penting, tapi juga apakah **provinsi miskin ikut mengejar provinsi kaya**.  

Dengan **σ-Convergence**, kita bisa menjawab:  
 *“Apakah pendidikan makin merata?”*  

**Hasilnya:** Ya, cukup merata sejak 2013, meski masih ada sedikit lonjakan ketimpangan di tahun 2024.


### Jadi kesimpulan akhirnya: 

Ketimpangan rata-rata lama sekolah antar provinsi di Indonesia cenderung menurun sejak 2010, menunjukkan adanya konvergensi pendidikan. Walau demikian, data 2024 memberi sinyal bahwa masih ada tantangan pemerataan di beberapa provinsi tertentu.

### **2. β-Convergence**

Regresi pertumbuhan terhadap level awal.
Misalnya: hitung pertumbuhan 2010–2024 → regresikan terhadap level awal 2010.

```python
# === β-Convergence: growth 2010–2024 terhadap level awal 2010 ===
# ----- 1) Siapkan data dasar hanya kolom yang dipakai
base = (
    df.loc[:, ["nama_provinsi", "tahun", "rata_rata_lama_sekolah"]]
      .dropna(subset=["nama_provinsi", "tahun", "rata_rata_lama_sekolah"])
      .copy()
)

# ----- 2) Ambil tahun awal & akhir (pastikan ada di data)
YEAR0, YEAR1 = 2010, 2024
d0 = base[base["tahun"] == YEAR0].rename(columns={"rata_rata_lama_sekolah": f"RLS_{YEAR0}"})
d1 = base[base["tahun"] == YEAR1].rename(columns={"rata_rata_lama_sekolah": f"RLS_{YEAR1}"})

# ----- 3) Gabungkan by provinsi, simpan hanya provinsi yang punya keduanya
merged = (
    pd.merge(d0[["nama_provinsi", f"RLS_{YEAR0}"]],
             d1[["nama_provinsi", f"RLS_{YEAR1}"]],
             on="nama_provinsi", how="inner")
)

# ----- 4) Bersihkan baris bermasalah: nol/negatif/NaN
mask_valid = (
    merged[f"RLS_{YEAR0}"].gt(0) &
    merged[f"RLS_{YEAR1}"].gt(0) &
    merged[[f"RLS_{YEAR0}", f"RLS_{YEAR1}"]].notna().all(axis=1)
)
bad = merged.loc[~mask_valid]
if not bad.empty:
    print("  data yang row-nya kosong - > Dibuang (nol/NaN/negatif):")
    display(bad)

merged = merged.loc[mask_valid].copy()

# ----- 5) Hitung growth tahunan rata-rata (continuous/CAGR log)
T = YEAR1 - YEAR0
merged["growth"] = (np.log(merged[f"RLS_{YEAR1}"]) - np.log(merged[f"RLS_{YEAR0}"])) / T

# ----- 6) Regresi: growth ~ level awal
X = sm.add_constant(merged[f"RLS_{YEAR0}"])
y = merged["growth"]
model = sm.OLS(y, X).fit(cov_type="HC1")   # robust SE (White/HC1)

print(model.summary())

# ----- 7) Visual bantu: scatter + garis regresi
plt.figure()
plt.scatter(merged[f"RLS_{YEAR0}"], merged["growth"])
# garis regresi
x_line = np.linspace(merged[f"RLS_{YEAR0}"].min(), merged[f"RLS_{YEAR0}"].max(), 100)
y_line = model.params["const"] + model.params[f"RLS_{YEAR0}"] * x_line
plt.plot(x_line, y_line)
plt.xlabel(f"Rata-rata Lama Sekolah {YEAR0}")
plt.ylabel("Pertumbuhan Tahunan (log) 2010–2024")
plt.title("β-convergence: growth vs level awal")
plt.show()

# ----- 8) Cek cepat arah konvergensi
beta = model.params.get(f"RLS_{YEAR0}")
sig  = model.pvalues.get(f"RLS_{YEAR0}")
print(f"\nKoefisien awal (β): {beta:.4f} | p-value: {sig:.4f}")
if np.isfinite(beta) and sig is not None:
    if (beta < 0) and (sig < 0.05):
        print(" Indikasi β-convergence: provinsi dengan level awal lebih rendah tumbuh lebih cepat.")
    else:
        print("❕ Tidak ada bukti kuat β-convergence (atau arahnya tidak negatif).")
```

**Apa yang Ditunjukkan Hasil Regresi?**

- **Koefisien RLS_2010 (β) = -0.0037**  
  Artinya, semakin rendah rata-rata lama sekolah di tahun 2010, semakin tinggi pertumbuhan rata-rata lama sekolah selama 2010–2024.  
  Nilai negatif inilah bukti adanya **konvergensi**.  

- **p-value = 0.007 (< 0.05)**  
  Koefisien ini signifikan secara statistik. Jadi, pola negatif tadi bukan kebetulan semata, melainkan punya dasar kuat dalam data.  

- **R-squared = 0.437**  
  Sekitar 43,7% variasi pertumbuhan bisa dijelaskan oleh level awal 2010.  
  Ini angka moderat, cukup masuk akal untuk data cross-provinsi.  

- **const = 0.0412**  
  Nilai ini bisa ditafsirkan sebagai *baseline growth* rata-rata tahunan, sebelum memperhitungkan level awal.  


**Apa yang Ditunjukkan Scatter Plot?**

- Titik-titik di scatter plot mewakili **provinsi**.  
- **Sumbu X**: rata-rata lama sekolah tahun 2010.  
- **Sumbu Y**: pertumbuhan tahunan (log) 2010–2024.  
- Garis regresi menurun → provinsi dengan level awal rendah tumbuh lebih cepat, sedangkan provinsi dengan level awal tinggi tumbuh lebih lambat.  
- **Contoh:** Papua atau NTT (awal rendah) tumbuh lebih pesat dibanding DKI Jakarta (awal tinggi).  


**Insight yang Bisa Diambil**

- **Ada β-Convergence**  
  Hasil ini mendukung hipotesis bahwa provinsi-provinsi yang tertinggal di awal periode 2010 berhasil mengejar ketertinggalannya sampai 2024.  

- **Pemerataan Pendidikan Semakin Nyata**  
  Kebijakan nasional seperti *Wajib Belajar 12 Tahun*, BOS, dan pembangunan infrastruktur pendidikan tampaknya membantu provinsi dengan titik awal rendah untuk tumbuh lebih cepat.  

- **Masih Ada Variasi yang Tidak Dijelaskan**  
  R-squared hanya 0.437, artinya masih ada faktor lain (misalnya kualitas guru, anggaran daerah, urbanisasi, ekonomi lokal) yang memengaruhi pertumbuhan pendidikan.  


**Alasan memakai Metode β-Convergence?**

**β-Convergence** adalah pendekatan yang melihat hubungan antara level awal dengan pertumbuhan.  

- Jika **koefisien negatif signifikan** → daerah yang lebih tertinggal tumbuh lebih cepat → ada **konvergensi**.  
- Metode ini populer dalam studi ekonomi regional (misalnya untuk PDB per kapita, IPM, pendidikan), karena memberi gambaran apakah *“yang tertinggal mengejar yang maju”*.  


## Kesimpulan

- Hasil regresi dan scatter menunjukkan indikasi kuat **β-Convergence** dalam rata-rata lama sekolah antar provinsi di Indonesia periode 2010–2024.  

- Dengan kata lain: provinsi yang awalnya tertinggal justru mengalami pertumbuhan pendidikan yang lebih cepat, sehingga ketimpangan antar provinsi berkurang.  

- Temuan ini memperkuat hasil **σ-Convergence** yang dihitung sebelumnya.  


## Gambaran Besar: σ-Convergence dan β-Convergence

- **σ-Convergence** → ketimpangan antar provinsi makin kecil.  
- **β-Convergence** → provinsi yang awalnya tertinggal memang tumbuh lebih cepat.
